## recalculate percentiles
#### conclusion: let's increase percentiles from median to 2stdev (0.85, 0.93, 0.99)

```python
filename = "../../Data/pipeline_matched_data_wo_linebreaks.pickle"
df = pd.read_pickle(filename)
```

```python
percentiles = [0.00135, 0.02275, 0.1587, 0.8413, 0.97725, 0.99865]
rows = ['0.1%', '2.3%', '15.9%', '50%', '84.1%', '97.7%', '99.9%']
df[['similarity']].describe(percentiles).loc[rows].round(3)
```

* old results - similarity
  * 0.1% --- 0.701
  * 2.3% --- 0.710
  * 15.9% -- 0.758
  * 50% ---- 0.847
  * 84.1% -- 0.922
  * 97.7% -- 0.963
  * 99.9% -- 0.983

```python
# use 0.99 instead of 1.00
low_p, med_p, high_p, v_high_p = 0.76, 0.85, 0.92, 0.99

print(f'\ntotal rows = {df.shape[0]}')
print('\n< low, med, high')
print(
    df[df.similarity < low_p].shape[0],
    df[df.similarity < med_p].shape[0],
    df[df.similarity < high_p].shape[0],
    df[df.similarity < v_high_p].shape[0],
)
print('\n> low, med, high')
print(
    df[df.similarity > low_p].shape[0],
    df[df.similarity > med_p].shape[0],
    df[df.similarity > high_p].shape[0],
    df[df.similarity > v_high_p].shape[0],
)
```

```python
print('\n< low, med, high %')
print(
    round(df[df.similarity < low_p].shape[0] / df.shape[0], 2),
    round(df[df.similarity < med_p].shape[0] / df.shape[0], 2),
    round(df[df.similarity < high_p].shape[0] / df.shape[0], 2),
    round(df[df.similarity < v_high_p].shape[0] / df.shape[0], 2),
)
print('\n> low, med, high %')
print(
    round(df[df.similarity > low_p].shape[0] / df.shape[0], 2),
    round(df[df.similarity > med_p].shape[0] / df.shape[0], 2),
    round(df[df.similarity > high_p].shape[0] / df.shape[0], 2),
    round(df[df.similarity > v_high_p].shape[0] / df.shape[0], 2),
)
```

```python
print(f"{round(df[df.similarity > v_high_p].shape[0] / df.shape[0], 3):0.1%} of entries have >99% similarity.")
```
