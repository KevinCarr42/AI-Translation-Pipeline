```python
import pandas as pd
```

## recalculate percentiles

```python
filename = "../../Data/pipeline_matched_data_wo_linebreaks.pickle"
df = pd.read_pickle(filename)
```

```python
percentiles = [0.00135, 0.02275, 0.1587, 0.8413, 0.97725, 0.99865]
rows = ['0.1%', '2.3%', '15.9%', '50%', '84.1%', '97.7%', '99.9%']
df[['similarity']].describe(percentiles).loc[rows].round(3)
```

## check quotes

```python
n = df[df.fr.str.contains('"')].shape[0] / df.shape[0]
print(f"percent of sentences with quotation marks = {n:.3%}")

n = df[df.fr.str.contains("'")].shape[0] / df.shape[0]
print(f"percent of sentences with single quote / apostrophe marks = {n:.3%}")
```

```python
n = df[df.fr.str.contains('“')].shape[0] / df.shape[0]
print(f"percent of sentences with curly quotation marks = {n:.3%}")

n = df[df.fr.str.contains("’")].shape[0] / df.shape[0]
print(f"percent of sentences with single curly quote / apostrophe marks = {n:.3%}")
```
