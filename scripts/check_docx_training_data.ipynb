```python
word_doc_training_data = Path("../../Data/worddoc_final_df.pickle")

df = pd.read_pickle(word_doc_training_data)
cols = [
    'OCR_issue', 'exclude_len_ratio', 'exclude_verb_ratio', 'exclude_noun_ratio',
    'exclude_entity_ratio', 'exclude_clause_ratio', 'exclude_figtext', 'exclude_too_short', 'has_date_refs',
    'unmatched_lexical_constraints', 'has_number_mismatch', 'exclude_wrong_language', 'is_duplicate'
]
df['n_exclusion_triggers'] = df[cols].sum(axis=1)
```
